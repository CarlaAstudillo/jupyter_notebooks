```python
prostitution_by_sex.columns = ['ORI','FEMALE_2016','FEMALE_2012','FEMALE_2013','FEMALE_2014', 'FEMALE_2015','MALE_2016','MALE_2012','MALE_2013','MALE_2014', 'MALE_2015']
```

```python
col_female_list = ['FEMALE_2016',
 'FEMALE_2012',
 'FEMALE_2013',
 'FEMALE_2014',
 'FEMALE_2015']
col_male_list = ['MALE_2016',
 'MALE_2012',
 'MALE_2013',
 'MALE_2014',
 'MALE_2015']
prostitution_by_sex['TOTAL_FEMALE'] = prostitution_by_sex[col_female_list].sum(axis=1)
prostitution_by_sex['TOTAL_MALE'] = prostitution_by_sex[col_male_list].sum(axis=1)
```

```python
import numpy as np
prostitution_by_sex['TIMES_MORE_FEMALE'] = prostitution_by_sex['TOTAL_FEMALE'].divide(prostitution_by_sex['TOTAL_MALE'])
prostitution_by_sex = prostitution_by_sex.replace([np.inf, -np.inf], 0)
prostitution_by_sex = prostitution_by_sex[['ORI','TOTAL_FEMALE', 'TOTAL_MALE', 'TIMES_MORE_FEMALE']]
prostitution_by_sex_wagencies = pd.merge(key_agencies2 , prostitution_by_sex, how='right', left_on='ORI', right_on='ORI')
prostitution_by_sex_wagencies = prostitution_by_sex_wagencies.drop_duplicates(subset='ORI', keep="first")
prostitution_by_sex_wagencies.fillna(0)
prostitution_by_sex_wagencies['MORE_MALE_OR_FEMALE'] = np.where(prostitution_by_sex_wagencies['TOTAL_FEMALE']>= prostitution_by_sex_wagencies['TOTAL_MALE'], 'MORE FEMALE', 'MORE MALE')
prostitution_by_sex_wagencies['TIE?'] = np.where(prostitution_by_sex_wagencies['TOTAL_FEMALE']== prostitution_by_sex_wagencies['TOTAL_MALE'], 'YES', 'NO')
prostitution_by_sex_wagencies['TOTAL'] = prostitution_by_sex_wagencies['TOTAL_FEMALE'] + prostitution_by_sex_wagencies['TOTAL_MALE'] 
prostitution_by_sex_wagencies['PCT_FEMALE'] = (prostitution_by_sex_wagencies['TOTAL_FEMALE']/prostitution_by_sex_wagencies['TOTAL'])*100
prostitution_by_sex_wagencies['PCT_MALE'] = (prostitution_by_sex_wagencies['TOTAL_MALE']/prostitution_by_sex_wagencies['TOTAL'])*100





prostitution_by_sex_sorted = prostitution_by_sex_wagencies.sort_values(by='TIMES_MORE_FEMALE', ascending=[False])
```
